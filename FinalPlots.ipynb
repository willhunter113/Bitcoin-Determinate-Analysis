```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-talk')
```

```python
file = 'NasdaqTrainResults.csv'
Nastrain = pd.read_csv(file, parse_dates = ['Date'], index_col = "Date")
file2 = 'NasdaqTestResults.csv'
Nastest = pd.read_csv(file2, parse_dates = ['Date'], index_col = "Date")
file3 = 'BitcoinTrainResults.csv'
Bittrain = pd.read_csv(file3, parse_dates = ['Date'], index_col = "Date")
file4 = 'BitcoinTestResults.csv'
Bittest = pd.read_csv(file4, parse_dates = ['Date'], index_col = "Date")
file5 = 'SPTrainResults.csv'
SPtrain = pd.read_csv(file5, parse_dates = ['Date'], index_col = "Date")
file6 = 'SPTestResults.csv'
SPtest = pd.read_csv(file6, parse_dates = ['Date'], index_col = "Date")
file7 = 'DowTrainResults.csv'
Dowtrain = pd.read_csv(file7, parse_dates = ['Date'], index_col = "Date")
file8 = 'DowTestResults.csv'
Dowtest = pd.read_csv(file8, parse_dates = ['Date'], index_col = "Date")
```

```python
with plt.style.context('bmh'):
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, sharex=True, figsize=(10,10))
    ax1.plot(Bittest, label='Bitcoin')
    ax2.plot(Nastest, label = "Nasdaq")
    ax3.plot(SPtest, label = "S&P 500")
    ax4.plot(Dowtest, label = "Dow Jones")
    fig.suptitle("Test Set Predictions vs. Actual Values", fontsize = 16)
    ax1.set_title('Bitcoin', fontsize = 12)
    ax2.set_title("Nasdaq", fontsize = 12)
    ax3.set_title("S&P 500", fontsize = 12)
    ax4.set_title("Dow Jones", fontsize = 12)
    ax4.legend(Bittest)

fig.savefig('TestPlot.png')
```

```python
with plt.style.context('bmh'):
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, sharex=True, figsize=(10,10))
    ax1.plot(Bittrain, label='Bitcoin')
    ax2.plot(Nastrain, label = "Nasdaq")
    ax3.plot(SPtrain, label = "S&P 500")
    ax4.plot(Dowtrain, label = "Dow Jones")
    fig.suptitle("Training Set Predictions vs. Actual Values", fontsize = 16)
    ax1.set_title('Bitcoin', fontsize = 12)
    ax2.set_title("Nasdaq", fontsize = 12)
    ax3.set_title("S&P 500", fontsize = 12)
    ax4.set_title("Dow Jones", fontsize = 12)
    ax1.legend(Bittrain)
fig.savefig('TrainPlot.png')
```
